```python
import pandas as pd
import json
df = pd.read_json('nodes.json')
```

```python

df['features'][0]
```

```python

df_clean = pd.DataFrame(list(df['features']))
df_clean
```

```python
df1 = pd.DataFrame(list(df_clean['properties']))[['amenity','denomination','name','religion']]
df1
```

```python
df2= pd.DataFrame(list(df_clean['geometry']))['coordinates']
df3 = pd.DataFrame()

lonlist = []
latlist = []

for row in df2:
    lonlist.append(row[0])
    latlist.append(row[1])


df3['lon'] = lonlist
df3['lat'] = latlist

df3
```

```python
df4 = pd.DataFrame(df_clean['id'])

for i,row in enumerate(df4['id']):
    df4['id'][i] = row.strip('node/')

df4
#df4['id'].strip('node/')
#.strip('node/')
```

```python
df1['id'] = df4['id']
df1[['lon','lat']] = df3[['lon','lat']]
```

```python
df1 = df1.dropna()
df1
```

```python
import re

cleanvals = []

for i,row in enumerate(df1['name']):
    
    if(row.isascii() == False):
        if (row == 'Chùa Bồ Đề'):
            cleanvals.append('Chua Bo De')
        else:
            #df1['name'][i] = re.sub(r'[^\x00-\x7F]+',"'", row)
            cleanvals.append(re.sub(r'[^\x00-\x7F]+',"'", row))
            #print(row)
            #print(re.sub(r'[^\x00-\x7F]+',"'", row))
        
    else:
        cleanvals.append(row)
        
```

```python
df1['name'] = cleanvals
```

```python
for i,row in enumerate(df1['name']):
    if(row.isascii() == False):
        print(row)
```
